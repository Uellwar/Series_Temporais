## Análise 80/20

```python
# Quais tipos de produtos contribuem com 80% das vendas

# Somatória do total de vendas liquidas por grupo
pivot_1 = ordem_vendas.groupby(["grupo_do_produto"])[["total_vendas_liquida"]].agg("sum")
# A divisão entre o total de vendas de cada grupo pela somatoria de todos os grupos (proporção)
pivot_1["proporcao_vendas"] = (pivot_1["total_vendas_liquida"]/pivot_1["total_vendas_liquida"].sum())
# multiplicando a porporcao *100 para pegar a porcentagem
pivot_1["porcentagem_vendas"] = round( (pivot_1["proporcao_vendas"])*100, 2)
# Ordenando os valores de proporcao de vendas para decrescente (os maiores acima)
pivot_1.sort_values("proporcao_vendas", ascending = False, inplace = True)
# Aplicando uma soma cumulativa na coluna proporcao de vendas. Aplicando uma funcao anonima para arredondar os valores para duas casas decimais e multiplicar *100
pivot_1["proporcao_vendas_acumulada"] = pivot_1["proporcao_vendas"].cumsum().apply(lambda x: "%.2f" % round(100*x,2))

# A media do total de vendas liquidas agrupado por grupo
pivot_2 = ordem_vendas.groupby(["grupo_do_produto"])[["total_vendas_liquida"]].agg("mean")
# Arredondando os valores para duas casas decimais
pivot_2["media_vendas_por_grupo"] = pivot_2["total_vendas_liquida"].apply(lambda x: "%.2f" % round(x,2))
# Renomeando a coluna para um nome que melhor descreve os dados
pivot_2.rename(columns = {"total_vendas_liquida":"media_vendas_por_grupo"}, inplace = True)

# Contagem de vendas agrupado por grupos
pivot_3 = ordem_vendas.groupby(["grupo_do_produto"])[["total_vendas_liquida"]].agg("count")
# Renomeando a coluna para um nome que melhor descreve os dados
pivot_3.rename(columns = {"total_vendas_liquida":"soma_total_de_pedidos"}, inplace = True)

# Realizando as mesclagens
pivot_1 = pivot_1.join(pivot_2)
pivot_1 = pivot_1.join(pivot_3)
```

```python
# Exportando a tabela para CSV para acessar pelo PowerBI
pivot_1.to_csv('/content/drive/MyDrive/Uelliton/vendas_online_2017_2019/dados_tratados/80_20.csv')
```
